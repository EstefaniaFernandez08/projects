```python
# filter dnb where 'BUSINESS_GROUP' is 'IPC'
dnb_ipc = dnb_master[dnb_master['BUSINESS_GROUP'] == 'IPC']

# filter dnb where 'BUSINESS_GROUP' is not 'IPC'
dnb_non_ipc = dnb_master[dnb_master['BUSINESS_GROUP'] != 'IPC']

# Display the shapes
print(f"D&B IPC only shape:{dnb_ipc.shape} and distinct normalized: {dnb_ipc['Business Name'].nunique()}")
print(f"D&B Remaining shape:{dnb_non_ipc.shape} and distinct normalized: {dnb_non_ipc['Business Name'].nunique()}") 
```

```python
# Group by 'SUPPLIER_NBR' and aggregate to find the range of 'Confidence Code' for each group
confidence_range = dnb_ipc.groupby('Business Name')['Confidence Code'].agg(['min', 'max'])

# Check if there are any differences in 'Confidence Code' within the duplicates
confidence_range['difference'] = confidence_range['max'] - confidence_range['min']

# Filter to see only those with differences
differences = confidence_range[confidence_range['difference'] > 0]

# Print the result
print("Groups with varying Confidence Codes:")
print(differences)

# Optionally, check how many groups have varying confidence codes
print(f"Number of supplier numbers with varying Confidence Codes: {len(differences)}")
```

```python

# Sort the DataFrame by 'Confidence Code' in descending order
dnb_ipc = dnb_ipc.sort_values(by='Confidence Code', ascending=False)

# Drop duplicates based on 'SUPPLIER_NBR', keeping the entry with the highest 'Confidence Code'
dnb_ipc = dnb_ipc.drop_duplicates(subset='Business Name')

# Print the length of the DataFrame after dropping duplicates
print(f"Length of DataFrame after dropping duplicates: {len(dnb_ipc)}")
```

```python
dnb_master_aft = pd.concat([dnb_ipc,dnb_non_ipc], ignore_index=True)

print(f"D&B IPC only shape:{dnb_master_aft.shape} and distinct normalized: {dnb_master_aft['Business Name'].nunique()}")
```
