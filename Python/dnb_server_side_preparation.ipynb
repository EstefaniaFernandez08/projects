```python
columns_to_select = [
    'SUPPLIER_NBR', 'SUPPLIER_ERP', 'SUPPLIER_ADDRESS', 'SUPPLIER_CITY',
    'SUPPLIER_STATE', 'SUPPLIER_ZIP_POSTAL_CDE', 'SUPPLIER_COUNTRY',
    'SUPPLIER_TAX_ID', 'SOURCE', 'Confidence Code', 'D-U-N-S Number',
    'Business Name', 'Global Ultimate Business Name'
] 

# Creating a new DataFrame with the selected columns
dnb_server_side = dnb_master[columns_to_select].copy()

# Renaming columns
dnb_server_side.rename(columns={
    'Business Name':'SUPPLIER_NORMALIZED',
    'Global Ultimate Business Name':'SUPPLIER_PARENT'
}, inplace=True)

# Displaying the fist rows to check
print(dnb_server_side.head())
```

```python
# Function to check for '.0' at the end of values and count rows
def check_decimal_zero(data, column_name):
    # Checking if the column is of object type and contains '.0'
    mask = data[column_name].astype(str).str.endswith('.0')
    count_decimal_zero = mask.sum()
    print(f"Rows with '.0' at the end in {column_name}: {count_decimal_zero}")

    # Checking for mixed data types by examining the unique types in the column
    mixed_types = data[column_name].apply(type).value_counts()
    print(f"Data types present in {column_name}:")
    print(mixed_types)

    # Check for duplicates where '.0' might cause inconsistency
    duplicates_count = data[column_name].duplicated().sum()
    print(f"Duplicate entries in {column_name}: {duplicates_count}")

# Columns to check
columns_to_check = [
    'SUPPLIER_NBR', 'SUPPLIER_ZIP_POSTAL_CDE', 'SUPPLIER_TAX_ID', 'D-U-N-S Number'
]

# Apply the function to each column
for column in columns_to_check:
    check_decimal_zero(dnb_server_side, column)
```

```python
from IPython.display import display

# Function to display duplicates or specific conditions, with sorting
def display_issues(data, column_name, condition):
    # Filtering based on the provided condition (duplicates or string condition)
    if condition == 'duplicates':
        duplicates_mask = data.duplicated(subset=[column_name], keep=False)
        filtered_data = data.loc[duplicates_mask, [column_name, 'SUPPLIER_ERP', 'SUPPLIER_NORMALIZED', 'SUPPLIER_PARENT']]
        # Sorting by the column of interest
        filtered_data.sort_values(by=[column_name], inplace=True)
    else:  # condition should be a string indicating a suffix like '.0'
        suffix_mask = data[column_name].astype(str).str.endswith(condition)
        filtered_data = data.loc[suffix_mask, [column_name, 'SUPPLIER_ERP', 'SUPPLIER_NORMALIZED', 'SUPPLIER_PARENT']]
        # Sorting by the column of interest
        filtered_data.sort_values(by=[column_name], inplace=True)
    
    print(f"Issues in {column_name} - {condition}")
    display(filtered_data)

# Displaying duplicate entries for specified columns
columns_to_inspect = ['SUPPLIER_NBR', 'SUPPLIER_ZIP_POSTAL_CDE', 'SUPPLIER_TAX_ID', 'D-U-N-S Number']
for column in columns_to_inspect:
    display_issues(dnb_server_side, column, 'duplicates')

# Displaying rows with '.0' at the end in 'D-U-N-S Number'
display_issues(dnb_server_side, 'D-U-N-S Number', '.0')

```

```python

# 1. Stripping '.0' from 'D-U-N-S Number' and 'SUPPLIER_TAX_ID'
dnb_server_side['D-U-N-S Number'] = dnb_server_side['D-U-N-S Number'].apply(lambda x: x.rstrip('.0') if isinstance(x, str) else x)
dnb_server_side['SUPPLIER_ZIP_POSTAL_CDE'] = dnb_server_side['SUPPLIER_ZIP_POSTAL_CDE'].apply(lambda x: x.rstrip('.0') if isinstance(x, str) else x)

# 2. Dropping duplicates in 'SUPPLIER_NBR' where 'SOURCE' is NaN
# First, temporarily fill NaNs in 'SOURCE' to preserve rows where 'SOURCE' is not NaN
dnb_server_side['temp_source'] = dnb_server_side['SOURCE'].fillna('temporary_fill')
# Now, drop duplicates based on 'SUPPLIER_NBR' considering only rows where original 'SOURCE' was NaN
dnb_server_side = dnb_server_side.drop_duplicates(subset=['SUPPLIER_NBR'], keep='first')
# Remove the temporary column
dnb_server_side.drop('temp_source', axis=1, inplace=True)

# 3. Replacing 'missing' values and NaNs with '' in 'SUPPLIER_ZIP_POSTAL_CDE' and 'SUPPLIER_TAX_ID'
dnb_server_side['SUPPLIER_ZIP_POSTAL_CDE'] = dnb_server_side['SUPPLIER_ZIP_POSTAL_CDE'].replace('missing', '').fillna('')
dnb_server_side['SUPPLIER_TAX_ID'] = dnb_server_side['SUPPLIER_TAX_ID'].replace('missing', '').fillna('')


# Displaying the cleaned DataFrame to check the first few rows
print(dnb_server_side.head())
```
