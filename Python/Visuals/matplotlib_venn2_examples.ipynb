```python
import pandas as pd
from matplotlib_venn import venn2
import matplotlib.pyplot as plt

# Assuming spend_final and dnb_master are already read into DataFrame
# spend_final = pd.read_csv('path_to_spend_final.csv')
# dnb_master = pd.read_csv('path_to_dnb_master.csv')

# Perform a full outer join on 'SUPPLIER_NBR' and align 'SUPPLIER_TYPE'
merged_df = pd.merge(spend_final, dnb_master, on=['SUPPLIER_NBR', 'SUPPLIER_TYPE'], how='outer', indicator=True)

# Count unique suppliers in each section
supplier_counts = merged_df.groupby(['_merge', 'SUPPLIER_TYPE'])['SUPPLIER_NBR'].nunique().unstack(fill_value=0)

# Plotting the Venn diagram
plt.figure(figsize=(8, 8))
venn_diagram = venn2(subsets=(1, 1, 1), set_labels=('Spend Final', 'DNB Master'))

# Update the labels with counts from supplier_counts
venn_diagram.get_label_by_id('10').set_text('\n'.join(f"{idx}: {val}" for idx, val in supplier_counts.loc['left_only'].items()))
venn_diagram.get_label_by_id('01').set_text('\n'.join(f"{idx}: {val}" for idx, val in supplier_counts.loc['right_only'].items()))
venn_diagram.get_label_by_id('11').set_text('\n'.join(f"{idx}: {val}" for idx, val in supplier_counts.loc['both'].items()))

plt.title('Venn Diagram of Supplier Number Overlap by Type')
plt.show()


```

```python
import pandas as pd
from matplotlib_venn import venn2
import matplotlib.pyplot as plt

# Assuming spend_final and dnb_master are already read into DataFrame
# spend_final = pd.read_csv('path_to_spend_final.csv')
# dnb_master = pd.read_csv('path_to_dnb_master.csv')

# Merge DataFrames with an indicator to identify intersections
merged_df = pd.merge(spend_final, dnb_master[['SUPPLIER_NBR']], on='SUPPLIER_NBR', how='left', indicator=True)

# Calculate the total spend for entries only in spend_final
total_spend_only_spend_final = merged_df[merged_df['_merge'] == 'left_only']['SPEND_USD'].sum()

# Calculate the total spend for entries that are in both spend_final and dnb_master
total_spend_both = merged_df[merged_df['_merge'] == 'both']['SPEND_USD'].sum()

# Plotting the Venn diagram
plt.figure(figsize=(8, 8))
venn_diagram = venn2(subsets=(total_spend_only_spend_final, 0, total_spend_both),
                     set_labels=('Spend Final', 'DNB Master'))

# Add annotations to the Venn diagram
venn_diagram.get_label_by_id('10').set_text(f"USD {total_spend_only_spend_final:,.0f}")
venn_diagram.get_label_by_id('11').set_text(f"USD {total_spend_both:,.0f}")

plt.title('Venn Diagram of Spending Overlap')
plt.show()

```
