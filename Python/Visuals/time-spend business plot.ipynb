```python
# Read the data into dataframes and print the each shape
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker

faf_data = pd.read_csv('fiscal_calendar_faf 5.6.24.csv', encoding='UTF-8-SIG')
sub_data = pd.read_csv('fiscal_calendar_sub 5.6.24.csv', encoding='UTF-8-SIG')

faf_shape = faf_data.shape
sub_shape = sub_data.shape

print('FAF shape:', faf_shape)
print('SUB shape:', faf_shape)
```

```python

# Define the formatter function
formatter = mticker.FuncFormatter(lambda x, pos: f'${x / 1e6}M' if x >= 1e6 else f'${x / 1e3}K')


# Define colors for the plot
colors = ['lightgray', 'skyblue']  # Adjust based on the number of years in your data

# Create the plot
plt.figure(figsize=(20, 7))

# Plot the data for each dataframe
plt.plot(faf_data['FISCAL_MONTH'], faf_data['TOTAL_SPEND_USD'], label='FAF', marker = 'o', linestyle='-', color=colors[0], linewidth=2.5)
plt.plot(sub_data['FISCAL_MONTH'], sub_data['TOTAL_SPEND_USD'], label='SUB', marker = 'o', linestyle='--', color=colors[1], linewidth=2.5)

# Add horizontal grid lines at each y-tick
plt.gca().yaxis.grid(True)

# Rotate x-axis labels to the left for better readability
plt.xticks(rotation=45)

# Set the y-axis formatter
plt.gca().yaxis.set_major_formatter(formatter)

# Find the maximum value to set the y-axis limit
max_value = max(faf_data['TOTAL_SPEND_USD'].max(), sub_data['TOTAL_SPEND_USD'].max())

# Set y-axis to range from 0 to the max value rounded to the nearest million or thousand
if max_value >= 1e6:
    plt.ylim(0, round(max_value / 1e6) * 1e6)
else:
    plt.ylim(0, round(max_value / 1e3) * 1e3)

# Labeling the axes
plt.title('Fiscal Calendar Spending Per Business Group')
plt.xlabel('Fiscal Month')
plt.ylabel('Total Spend (USD)')

# Adding a legend
plt.legend()

# Display the plot
plt.show()
```
