```python
import re

# Function to replace special characters, handle whitespaces, and convert to uppercase
def replace_char(word):
    # Remove special characters
    special_chars = ['$', '%', '^', '*', '#', '(', ')', '[', ']', ';', ':', '"', "'", ',', '.']
    for char in special_chars:
        word = word.replace(char, '')
    
    # Remove leading and trailing spaces, and replace multiple spaces with a single space
    word = re.sub(r'\s+', ' ', word.strip())
    
    # Convert to uppercase
    return word.upper()

# Apply normalization function to the DataFrame columns
dnb_master['SUPPLIER_NORMALIZED'] = dnb_master['SUPPLIER_NORMALIZED'].apply(replace_char)
dnb_master['SUPPLIER_PARENT'] = dnb_master['SUPPLIER_PARENT'].apply(replace_char)
```

## step 3 - Join & Concatenate the Results for a Complete Suppplier Master

### step 3.1 - To update the supplier_master with the most recent normalized values from the dnb_master and set the source status.

```python

# Perform a left join to update supplier_master with the most recent data from dnb_master
merged_df = pd.merge(supplier_master, dnb_master[['SUPPLIER_NBR', 'SUPPLIER_NORMALIZED', 'SUPPLIER_PARENT']],
                     on='SUPPLIER_NBR', how='left', suffixes=('', '_updated'))

# Update the 'SUPPLIER_NORMALIZED' and 'SUPPLIER_PARENT' fields in supplier_master
supplier_master['SUPPLIER_NORMALIZED'] = merged_df['SUPPLIER_NORMALIZED_updated'].combine_first(merged_df['SUPPLIER_NORMALIZED'])
supplier_master['SUPPLIER_PARENT'] = merged_df['SUPPLIER_PARENT_updated'].combine_first(merged_df['SUPPLIER_PARENT'])

# Set the 'SOURCE_STATUS' to distinguish between shared and exclusive entries
supplier_master['SOURCE_STATUS'] = merged_df['SUPPLIER_NORMALIZED_updated'].apply(lambda x: 'Shared' if pd.notna(x) else 'Supplier Exclusive')

# Output the final DataFrame information
print(f"Supplier master shape: {supplier_master.shape}")
```

### step 3.2 - To ensure the final DataFrame (final_df) preserves the structure of supplier_master while incorporating the exclusive entries from dnb_master

```python
# Identify DNB exclusive entries
exclusive_dnb = dnb_master[~dnb_master['SUPPLIER_NBR'].isin(supplier_master['SUPPLIER_NBR'])]

# Preparing DNB exclusive entries to match the supplier_master structure
columns_needed = list(supplier_master.columns)  # Ensuring these are the only columns we use
exclusive_dnb = exclusive_dnb.reindex(columns=columns_needed)  # Reindexing to match exactly
exclusive_dnb = exclusive_dnb.assign(**{col: pd.NA for col in columns_needed if col not in exclusive_dnb.columns})
exclusive_dnb['SOURCE_STATUS'] = 'DNB Exclusive'  # Adding status

# Concatenate the adjusted exclusive_dnb to the supplier_master
final_df = pd.concat([supplier_master, exclusive_dnb], ignore_index=True)

# Output the final DataFrame information
print(f"Final result shape: {final_df.shape}")
```
