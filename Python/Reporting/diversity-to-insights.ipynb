```python

# Summarize SPEND_USD for every distinct SUPPLIER_NBR in spend_final
spend_summary = spend_final.groupby('SUPPLIER_NBR', as_index=False).agg({'SPEND_USD': 'sum'})

# Prepare spend_final with distinct suppliers
spend_final_distinct = spend_summary.merge(
    spend_final.drop(columns=['SPEND_USD']).drop_duplicates('SUPPLIER_NBR'),
    on='SUPPLIER_NBR', how='left'
)

# Prepare diversity_final with distinct suppliers
diversity_final_distinct = diversity_final.drop(columns=['SPEND_USD']).drop_duplicates('SUPPLIER_NBR')

# Merge the two distinct DataFrames
merged_df = pd.merge(spend_final_distinct, diversity_final_distinct, on='SUPPLIER_NBR', how='outer', suffixes=('_spend', '_diversity'))

# Define the columns for the final DataFrame with their respective sources
columns_from_diversity = [
    'DIVERSITY_FLAG', 'SD_8A_SRC', 'SD_8A_SRC_VALID', 'SD_ABILITY_ONE_SRC', 'SD_ABILITY_ONE_SRC_VALID',
    'SD_ACDBE_SRC', 'SD_ACDBE_SRC_VALID', 'SD_ALT_MBE_SRC', 'SD_ALT_MBE_SRC_VALID', 'SD_ALT_WBE_SRC',
    'SD_ALT_WBE_SRC_VALID', 'SD_ANC_INDIAN_TRIBE_SRC', 'SD_ANC_SRC_VALID', 'SD_CAGE', 'SD_CORP_STRUCTURE',
    'SD_DBE_SRC', 'SD_DBE_SRC_VALID', 'SD_DISABILITYIN_CERTIFIED', 'SD_DISABLED_SRC', 'SD_DISABLED_SRC_VALID',
    'SD_DIVERSITY_ATTRIBUTES', 'SD_EDWOSB_SRC', 'SD_EDWOSB_SRC_VALID', 'SD_ETHNICITY', 'SD_HBCU_SRC',
    'SD_HBCU_SRC_VALID', 'SD_HUB_SRC', 'SD_HUB_SRC_VALID', 'SD_INDUSTRY', 'SD_IS_8A', 'SD_IS_ABILITY_ONE',
    'SD_IS_ACDBE', 'SD_IS_ANC_INDIAN_TRIBE', 'SD_IS_DBE', 'SD_IS_DISABLED', 'SD_IS_EDWOSB', 'SD_IS_HBCU',
    'SD_IS_HUB', 'SD_IS_LGBT', 'SD_IS_MBE', 'SD_IS_SDB', 'SD_IS_SDVET', 'SD_IS_SELF_SDB', 'SD_IS_SMALL',
    'SD_IS_VET', 'SD_IS_WBE', 'SD_IS_WOSB', 'SD_LGBT_SRC', 'SD_LGBT_SRC_VALID', 'SD_MBE_SRC', 'SD_MBE_SRC_VALID',
    'SD_NAICS', 'SD_NGLCC_CERTIFIED', 'SD_NMSDC_CERTIFIED', 'SD_NVBDC_CERTIFIED', 'SD_SAM_EXPDATE',
    'SD_SAM_REGISTERED', 'SD_SDB_SRC', 'SD_SDB_SRC_VALID', 'SD_SDVET_SRC', 'SD_SDVET_SRC_VALID', 'SD_SELFSDB_SRC',
    'SD_SELFSDB_SRC_VALID', 'SD_SIC', 'SD_SMALL_SRC', 'SD_SMALL_SRC_VALID', 'SD_VET_SRC', 'SD_VET_SRC_VALID',
    'SD_WBE_SRC', 'SD_WBE_SRC_VALID', 'SD_WBENC_CERTIFIED', 'SD_WOSB_SRC', 'SD_WOSB_SRC_VALID'
]

columns_from_spend = [
    'SUPPLIER_NBR', 'FILE_NAME', 'FISCAL_YEAR', 'FISCAL_QUARTER', 'LS_CAT1', 'LS_CAT2', 'LS_CAT3', 'SPEND_USD',
    'SUPPLIER_ADDRESS_1', 'SUPPLIER_ADDRESS_2', 'SUPPLIER_CITY', 'SUPPLIER_COUNTRY', 'SUPPLIER_ERP',
    'SUPPLIER_NORMALIZED', 'SUPPLIER_STATE', 'SUPPLIER_ZIP_POSTAL_CDE', 'BUSINESS_GROUP', 'SUPPLIER_TYPE',
    'FPA_FUNCTION', 'HOLDING_COMPANY'
]

# Initialize the final DataFrame with the specified columns and order
final_df = pd.DataFrame(columns=list(columns.keys()))

# Assign fields from diversity_final to final_df
for field in columns_from_diversity:
    diversity_field = f"{field}_diversity"
    if diversity_field in merged_df.columns:
        final_df[field] = merged_df[diversity_field]
    elif field in merged_df.columns:  # Handle the case where there is no suffix
        final_df[field] = merged_df[field]

# Assign fields from spend_final to final_df
for field in columns_from_spend:
    spend_field = f"{field}_spend"
    if spend_field in merged_df.columns:
        final_df[field] = merged_df[spend_field]
    elif field in merged_df.columns:  # Handle the case where there is no suffix
        final_df[field] = merged_df[field]

# Display the final DataFrame
print("Final DataFrame:")
print(final_df.shape)
```

```python
# Define columns for spend-related fields
spend_fields = [
    'SUPPLIER_NBR', 'FILE_NAME', 'FISCAL_YEAR', 'FISCAL_QUARTER', 'LS_CAT1', 'LS_CAT2', 'LS_CAT3', 'SPEND_USD',
    'SUPPLIER_ADDRESS_1', 'SUPPLIER_ADDRESS_2', 'SUPPLIER_CITY', 'SUPPLIER_COUNTRY', 'SUPPLIER_ERP',
    'SUPPLIER_NORMALIZED', 'SUPPLIER_STATE', 'SUPPLIER_ZIP_POSTAL_CDE', 'BUSINESS_GROUP', 'SUPPLIER_TYPE',
    'FPA_FUNCTION', 'HOLDING_COMPANY'
]

# Filter for suppliers in the diversity report (United States only)
diversity_suppliers_df = final_df[
    final_df['SUPPLIER_NBR'].isin(diversity_final_distinct['SUPPLIER_NBR']) & 
    (final_df['SUPPLIER_COUNTRY'] == 'UNITED STATES')
]

# Export to CSV for Oracle upload
diversity_suppliers_df.to_csv('SUB_FY24-Q1_diversity_refresh_5.21.24.csv', index=False)

# Filter for exclusions: SUPPLIER_TYPE = 'EMPLOYEE' and SUPPLIER_COUNTRY is not 'UNITED STATES'
exclusions_df = final_df[
    (final_df['SUPPLIER_TYPE'] == 'EMPLOYEE') | 
    (final_df['SUPPLIER_COUNTRY'] != 'UNITED STATES')
]

# Print value counts for exclusions
print("Exclusions - Supplier Type Counts:\n", exclusions_df['SUPPLIER_TYPE'].value_counts())
print("Exclusions - Supplier Country Counts:\n", exclusions_df['SUPPLIER_COUNTRY'].value_counts())

# Export exclusions to CSV
exclusions_df[spend_fields].to_csv('diversity_external_exclusions_5.21.24.csv', index=False)

# Filter for remaining suppliers not in the diversity report or exclusions
remaining_suppliers_df = final_df[
    (~final_df['SUPPLIER_NBR'].isin(diversity_final_distinct['SUPPLIER_NBR'])) &
    ~((final_df['SUPPLIER_TYPE'] == 'EMPLOYEE') | (final_df['SUPPLIER_COUNTRY'] != 'UNITED STATES'))
]

# Select only spend-related fields
remaining_spend_fields_df = remaining_suppliers_df[spend_fields]

# Export remaining suppliers for resubmission to CSV
remaining_spend_fields_df.to_csv('sub_fy24-q1_sio_resubmission_5.21.24.csv', index=False)

# Print row counts for verification
print("Exports completed successfully.")
print(f"Suppliers for Oracle: {diversity_suppliers_df.shape}")
print(f"Suppliers for Exclusions: {exclusions_df.shape}")
print(f"Suppliers for SIO resubmission: {remaining_spend_fields_df.shape}")

```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib_venn import venn3

# Assuming final_df, diversity_suppliers_df, exclusions_df, and remaining_suppliers_df are already defined

# Calculate SPEND_USD for each category
spend_diversity = diversity_suppliers_df['SPEND_USD'].sum()
spend_exclusions = exclusions_df['SPEND_USD'].sum()
spend_remaining = remaining_suppliers_df['SPEND_USD'].sum()

# Calculate grand total SPEND_USD
total_spend = spend_diversity + spend_exclusions + spend_remaining

# Calculate percentages for each category and round them
percent_diversity = round((spend_diversity / total_spend) * 100, 2)
percent_exclusions = round((spend_exclusions / total_spend) * 100, 2)
percent_remaining = round((spend_remaining / total_spend) * 100, 2)

# Calculate row counts for each category
count_diversity = diversity_suppliers_df.shape[0]
count_exclusions = exclusions_df.shape[0]
count_remaining = remaining_suppliers_df.shape[0]

# Add submission status to diversity suppliers
diversity_suppliers_df['SUBMITTED'] = 'Yes'

# Remaining suppliers are not submitted
remaining_suppliers_df['SUBMITTED'] = 'No'

# Optionally, export these tagged dataframes to check them
diversity_suppliers_df.to_csv('diversity_suppliers_with_submission_status.csv', index=False)
remaining_suppliers_df.to_csv('remaining_suppliers_with_submission_status.csv', index=False)

# Create Venn diagrams
plt.figure(figsize=(12, 6))

# Venn diagram for SPEND_USD as percentages
plt.subplot(1, 2, 1)
venn3(
    subsets=(percent_diversity, percent_exclusions, percent_remaining, 0, 0, 0, 0),
    set_labels=('Diversity', 'Exclusions', 'Remaining'),
    set_colors=('skyblue', 'lightgray', 'lightgray')
)
plt.title('Spend USD Distribution (%)')

# Venn diagram for row counts
plt.subplot(1, 2, 2)
venn3(
    subsets=(count_diversity, count_exclusions, count_remaining, 0, 0, 0, 0),
    set_labels=('Diversity', 'Exclusions', 'Remaining'),
    set_colors=('skyblue', 'lightgray', 'lightgray')
)
plt.title('Supplier Number Distribution')

plt.tight_layout()
plt.show()

```
